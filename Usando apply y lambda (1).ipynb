# Funciones Lambda
En Python, se usa  la palabra clave __lambda__ para declarar una función anónima, por lo que nos referimos a ellas como "funciones lambda". Una función anónima se refiere a una función declarada sin nombre. Aunque sintácticamente se ven diferentes, las funciones lambda se comportan de la misma manera que las funciones regulares que se declaran usando la palabra clave __def__

Las funciones lambda se utilizan cuando se necesita una función por un período corto de tiempo.

```python
def producto(x,y):
    producto=x*y
    return(producto)

producto(5,2)
```

```python
producto = lambda x, y : x * y
producto(5,3)
```

```python
# Crea una nueva columna con el nombre 'prom_petalo', la cual tiene el promedio de las medidas del ancho y longitud del petalo
iris = iris.assign(
prom_petalo= lambda x: (x.petalo_longitud+x.petalo_ancho)/2)
iris.head(5)
```

```python
# Crea una nueva columna con nombre 'prom_sepalo' usando .apply, la cual contiene el promedio de las medidas del 
# ancho y longitud del sepalo utilizando apply
iris['prom_sepalo'] = iris.apply(
lambda x: (x.sepalo_longitud+x.sepalo_ancho)/2, axis= 1)
iris.head(5)
```

```python
iris.head()
iris['prueba']=iris.petalo_longitud+2
iris.head()
```

```python
# borra una columna "in-place" (en el mismo marco de datos)
del iris['prom_sepalo']
iris.head(5)
```
